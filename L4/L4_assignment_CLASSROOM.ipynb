## Task 1: Dataset and split
Load a dataset and create a reproducible train/test split.

```python
# TODO: Load a dataset (iris, breast_cancer, or diabetes)
from sklearn.datasets import load_iris, load_diabetes


# I produktion:
# data = pd.load_csv("filnamn.csv")
# Den datorn kan dock behöva mer förarbete
# Med färdigbyggda dataset, behövs sällan preprocessing
# Men, EDA är fortfarande väldigt viktigt!
# Om man har bra ETL / data engineers, kan datan vara användbar från början




data = load_iris(as_frame=True)
print(data)

X = data.data
y = data.target
target_names = data.target_names

print("X: ", X)
print("y: ", y)



```

```python
# TODO: Create train/test split with a fixed random_state

# Here we often use a function (train_test_split)
# The funtion splits our data neatly into piles of train and test
# We could do this manually


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

```

# Ett EDA steg, vi undersöker datan med plottande

```python
# Sliceexempel

# lista[:] - hämtar all data
# alltså: [:] i en slice betyder, allt
X[:]

# Men, : kan modifieras, vi kan hämta all data innan
# eller efter
#[12:] hämtar all data från rad 12 och nedåt
X[12:]

# [:12] hämtar all data fram till rad 12
X[:12]

# Vi kan också slica i fler än en dimensioner
# Det gör vi genom att använda ett kommatkn
# som ett exp :  [[1,2,3]] , [4,5,6]]

# Vi testar
# Note, med en dataframe slicar vi med iloc istället
# för med bara hakparenteser [], men principen är 
# den samma.

X.iloc[:,0]
```

```python
import matplotlib.pyplot as plt


# Med plt.plot() kan vi plotta data
# plot() tar data på listform, som kan skicka in på
# olika sätt

# Med plt.plot() plottar vi ett linjediagram
# Linjediagramet har jättemånga inställningar 
# De kan ni gräva i när ni behöver plotta
# "o":et är ett exempel på en inställning
# Den gör att vi plottar punkter istället för linjer
# plt.plot(X,y, 'o')


# Vår data är högdimensionell (4 dimensioner), 
# Så vi väljer att plotta två dimensioner
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap="viridis", s=30)

plt.title("Iris: Sepal length vs sepal width")
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.tight_layout()
plt.show()

```
